```python
import pandas as pd
import numpy as np
import json
from faker import Faker
from datetime import datetime, timedelta
import random
import openpyxl

fake = Faker('fr_FR')
np.random.seed(42)
random.seed(42)

# ============================================
# 1. STATIONS MASTER (Excel)
# ============================================
stations = {
    'ID_Station': [f'ST{str(i).zfill(3)}' for i in range(1, 21)],
    'Nom_Station': [
        'Haouz-Nord', 'Haouz-Sud', 'Gharb-Est', 'Gharb-Ouest',
        'Souss-Côtier', 'Souss-Intérieur', 'Tadla-Centre', 'Tadla-Nord',
        'Doukkala-1', 'Doukkala-2', 'Moulouya-Amont', 'Moulouya-Aval',
        'Saiss-Principal', 'Saiss-Secondaire', 'Draa-Haut', 'Draa-Bas',
        'Tensift-Amont', 'Tensift-Aval', 'Loukkos-Nord', 'Loukkos-Sud'
    ],
    'Ville': [
        'Marrakech', 'Marrakech', 'Kenitra', 'Kenitra',
        'Agadir', 'Agadir', 'Beni Mellal', 'Beni Mellal',
        'El Jadida', 'El Jadida', 'Oujda', 'Oujda',
        'Fès', 'Fès', 'Ouarzazate', 'Ouarzazate',
        'Essaouira', 'Essaouira', 'Larache', 'Larache'
    ],
    'Altitude': [
        450, 520, 15, 25, 30, 180, 390, 420,
        20, 35, 950, 320, 580, 610, 1150, 780,
        45, 60, 10, 18
    ],
    'Zone_Geo': [
        'Haouz', 'Haouz', 'Gharb', 'Gharb',
        'Souss-Massa', 'Souss-Massa', 'Tadla', 'Tadla',
        'Doukkala', 'Doukkala', 'Moulouya', 'Moulouya',
        'Saiss', 'Saiss', 'Draa-Tafilalet', 'Draa-Tafilalet',
        'Tensift', 'Tensift', 'Loukkos', 'Loukkos'
    ],
    'Capteur_Type': [
        'Digital', 'Analogique', 'Digital', 'Digital',
        'Analogique', 'Digital', 'Digital', 'Analogique',
        'Digital', 'Analogique', 'Digital', 'Digital',
        'Analogique', 'Digital', 'Analogique', 'Digital',
        'Digital', 'Analogique', 'Digital', 'Digital'
    ]
}

df_stations = pd.DataFrame(stations)
df_stations.to_excel('../data/stations_master.xlsx', index=False)
print("✅ stations_master.xlsx créé :", df_stations.shape)

# ============================================
# 2. RELEVES METEO (CSV)
# ============================================
records = []
start_date = datetime(2024, 1, 1)

for i in range(5000):
    station = random.choice(stations['ID_Station'])
    timestamp = start_date + timedelta(hours=random.randint(0, 8760))
    
    # Valeurs normales
    temp = round(random.uniform(5, 45), 1)
    hum = round(random.uniform(10, 95), 1)
    wind = round(random.uniform(0, 80), 1)
    
    # Injection de valeurs aberrantes (5%)
    if random.random() < 0.05:
        temp = round(random.uniform(61, 75), 1)
    
    # Injection de valeurs manquantes (3%)
    if random.random() < 0.03:
        temp = None
    if random.random() < 0.03:
        hum = None

    records.append({
        'timestamp': timestamp.strftime('%Y-%m-%d %H:%M:%S'),
        'st_id': station,
        'temp_c': temp,
        'hum_pct': hum,
        'wind_speed': wind
    })

df_meteo = pd.DataFrame(records)
df_meteo.to_csv('../data/relevés_meteo.csv', index=False, encoding='utf-8')
print("✅ relevés_meteo.csv créé :", df_meteo.shape)

# ============================================
# 3. NOTIFICATIONS (JSON)
# ============================================
severity_levels = ['RAS', 'Jaune', 'Orange', 'Rouge']
precip_types = ['Pluie', 'Grêle', 'Neige']
alert_messages = {
    'RAS': 'Situation normale, aucune alerte.',
    'Jaune': 'Vigilance recommandée, précipitations modérées.',
    'Orange': 'Alerte orange : risque élevé de crues.',
    'Rouge': 'ALERTE ROUGE : danger immédiat, évacuation conseillée.'
}

notifications = []
start_date = datetime(2024, 1, 1)

for i in range(500):
    station = random.choice(stations['ID_Station'])
    date = start_date + timedelta(days=random.randint(0, 365))
    severity = random.choices(
        severity_levels,
        weights=[50, 25, 15, 10]
    )[0]

    notifications.append({
        'date': date.strftime('%Y-%m-%d'),
        'station_code': station,
        'precip_mm': round(random.uniform(0, 120), 1),
        'type_precip': random.choice(precip_types),
        'severity_index': severity,
        'alert_msg': alert_messages[severity]
    })

with open('../data/notifications.json', 'w', encoding='utf-8') as f:
    json.dump(notifications, f, ensure_ascii=False, indent=2)

print("✅ notifications.json créé :", len(notifications), "enregistrements")
print("\n🎉 Toutes les données sources sont générées !")
```
